```python
import whisper
import wikipedia
import webbrowser
import os

# ---------- Configurações ----------
wikipedia.set_lang("pt")  # Wikipedia em português

def speak(text):
    """No Colab, apenas imprime o texto"""
    print("Assistente:", text)

def listen_from_file(file_path):
    """Usa Whisper para transcrever áudio WAV em texto"""
    if not os.path.exists(file_path):
        speak("Não encontrei o arquivo de áudio.")
        return ""

    # Carrega modelo Whisper
    model = whisper.load_model("base")  # tiny, small, medium ou large
    result = model.transcribe(file_path, language="pt")
    command = result["text"].lower().strip()
    print("Você disse:", command)
    return command

def process_command(command):
    """Executa ações com base no comando de voz"""
    if "wikipedia" in command:
        topic = command.replace("wikipedia", "").strip()
        if topic:
            try:
                summary = wikipedia.summary(topic, sentences=2, auto_suggest=False)
                speak(summary)
            except:
                speak("Não encontrei resultados no Wikipedia.")
        else:
            speak("Diga o que deseja pesquisar no Wikipedia.")
    elif "youtube" in command:
        speak("Abrindo o YouTube")
        # No Colab não abre browser, mas podemos imprimir link
        print("Link:", "https://www.youtube.com")
    elif "farmácia" in command:
        speak("Procurando farmácia próxima")
        print("Link:", "https://www.google.com/maps/search/farmácia+próxima")
    else:
        speak("Desculpe, não entendi o comando.")

# ---------- Programa principal ----------
arquivo_audio = "Gravando.wav"  # nome do arquivo enviado

speak("Olá, sou sua assistente virtual.")

comando = listen_from_file(arquivo_audio)

if comando:
    process_command(comando)
```
